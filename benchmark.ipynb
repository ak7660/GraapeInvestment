```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the investment parameters
investment_amount = 100_000_000  # 100M USD
start_date = "2023-10-01"
end_date = pd.Timestamp.today().strftime('%Y-%m-%d')

# Fetch S&P 500 data
sp500 = yf.Ticker("^GSPC")
sp500_data = sp500.history(start=start_date, end=end_date)

# Calculate the total asset movement
sp500_data['Investment Value'] = (sp500_data['Close'] / sp500_data['Close'].iloc[0]) * investment_amount

# Calculate the returns
sp500_data['Returns'] = sp500_data['Close'].pct_change() * 100  # Convert to percentage

# Calculate the drawdown
sp500_data['Peak'] = sp500_data['Investment Value'].cummax()
sp500_data['Drawdown'] = (sp500_data['Investment Value'] - sp500_data['Peak']) / sp500_data['Peak'] * 100

# Plot the total asset movement
plt.figure(figsize=(14, 7))
plt.plot(sp500_data.index, sp500_data['Investment Value'], label='Total Asset Movement')
plt.title('Benchmark with S&P 500 (2023-24)')
plt.xlabel('Date')
plt.ylabel('Investment Value (USD)')
plt.legend()
plt.grid(True)
plt.show()

# Plot the returns
plt.figure(figsize=(14, 7))
plt.plot(sp500_data.index, sp500_data['Returns'], label='Daily Returns')
plt.title('Daily Returns of S&P 500')
plt.xlabel('Date')
plt.ylabel('Returns (%)')
plt.legend()
plt.grid(True)
plt.show()

# Plot the drawdown with filled color
plt.figure(figsize=(14, 7))
plt.plot(sp500_data.index, sp500_data['Drawdown'], label='Equity Drawdown', color='red')
plt.fill_between(sp500_data.index, sp500_data['Drawdown'], color='red', alpha=0.3)
plt.title('Equity Drawdown of S&P 500')
plt.xlabel('Date')
plt.ylabel('Drawdown (%)')
plt.legend()
plt.grid(True)
plt.show()
```
